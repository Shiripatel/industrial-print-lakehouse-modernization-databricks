This notebook derives customer / PSP-level performance metrics from the Gold fact tables.


```
from pyspark.sql.functions import avg, count

fact = spark.table("industrial_print.gold.fact_order_fulfillment")
psp = spark.table("industrial_print.gold.dim_psp")

psp_perf = (
    fact.join(psp, "psp_id")
        .groupBy("psp_id", "psp_name")
        .agg(
            count("order_id").alias("total_orders"),
            avg("hours_between_print_and_ship").alias("avg_hours_between_print_and_ship")
        )
)

psp_perf.display()
```
